# Task 1: Data Preparation and Cleaning
**Goal:** Clean the raw dataset by handling missing values, identifying duplicates, and correcting data formats using Python and Pandas.

### Steps to be performed:
1. Load the dataset and inspect its structure.
2. Identify and handle missing or null values.
3. Identify and remove duplicate rows.
4. Correct data formats (Dates, Text, and Numeric types).
5. Verify the cleaned dataset.

```python
from google.colab import files
uploaded = files.upload()
```

```python
# Import required libraries
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_excel('Dataset for Data Analytics.xlsx')

# Display the first 5 rows to see what we are working with
print("--- First 5 Rows of the Dataset ---")
df.head()
```

## Step 1: Identify Missing or Null Values
Let's check the structural overview of the dataset and count how many null values exist in each column.

```python
# Display dataset summary (data types and non-null counts)
print("--- Dataset Information ---")
df.info()

print("\n--- Missing Values Count Per Column ---")
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
```

## Step 2: Handle Missing Values
Depending on the column type, we will handle missing values logically:
* For important identifiers like `OrderID` or critical missing information, rows might be dropped if they are completely corrupted (though looking at our data, columns like `CouponCode` naturally contain nulls because not every customer uses a coupon).
* We will fill missing values in categorical data like `CouponCode` with a placeholder like `'NONE'`.

```python
# Handling missing data
# If 'CouponCode' is null, it means no coupon was used. Let's fill it with 'NONE'
if 'CouponCode' in df.columns:
    df['CouponCode'] = df['CouponCode'].fillna('NONE')

# Verify if any critical missing values remain
print("Remaining missing values per column:")
print(df.isnull().sum())
```

## Step 3: Identify and Remove Duplicates
Duplicate records can skew analytics results. We will identify any identical rows and remove them.

```python
# Check for exact duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows found: {duplicate_count}")

# If duplicates exist, drop them keeping the first occurrence
if duplicate_count > 0:
    df = df.drop_duplicates(keep='first')
    print("Duplicates removed successfully!")
else:
    print("No duplicates to remove.")
```

## Step 4: Correct Data Formats
To ensure accurate analysis, we need to correct data formats:
* Convert the **Date** column into a proper datetime format.
* Ensure **Numeric** columns (`Quantity`, `UnitPrice`, `TotalPrice`) are floats or integers.
* Ensure **Text** columns are uniformly stripped of accidental whitespaces.

```python
# 1. Correct Date Format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# 2. Correct Numeric Formats
df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')
df['UnitPrice'] = pd.to_numeric(df['UnitPrice'], errors='coerce')
df['TotalPrice'] = pd.to_numeric(df['TotalPrice'], errors='coerce')

# 3. Clean Text Columns (Strip leading/trailing whitespaces)
text_columns = ['Product', 'PaymentMethod', 'OrderStatus', 'ReferralSource']
for col in text_columns:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()

print("Data formats corrected! Let's check the updated column types:")
df.dtypes
```

## Step 5: Sanity Check & Export Cleaned Data
Now that our data is clean, no duplicates exist, and formats are correct, let's look at the final summary and save the file.

```python
# Final validation check
print(f"Final shape of the cleaned dataset: {df.shape}")

# Save the cleaned dataset to a new CSV file
df.to_csv('Cleaned_Data_Analytics_Dataset.csv', index=False)
print("Cleaned dataset successfully exported as 'Cleaned_Data_Analytics_Dataset.csv'!")
```

# Task 2: Exploratory Data Analysis (EDA)
**Goal:** Analyze the cleaned dataset to calculate basic statistics, identify overarching trends, discover outliers, and summarize key insights.

### Steps to be performed:
1. Compute descriptive statistics for numeric features.
2. Analyze sales trends over time.
3. Identify distribution patterns and potential outliers.
4. Summarize key behavioral observations (e.g., top products, popular payment methods).

```python
# Ensure Date column is in datetime format for accurate sorting
df['Date'] = pd.to_datetime(df['Date'])

print("Dataset ready for analysis!")
```

## Step 1: Calculate Basic Statistics
Let's look at the summary statistics (mean, median, standard deviation, min, max, and count) for our numeric attributes like `Quantity`, `UnitPrice`, and `TotalPrice`. This helps establish our baseline benchmarks.

```python
# Generate descriptive statistics
print("--- Summary Statistics for Numeric Columns ---")
summary_stats = df[['Quantity', 'UnitPrice', 'TotalPrice', 'ItemsInCart']].describe()

# Add median explicitly since .describe() includes 50% (which is the median), let's make it clear:
print(summary_stats)

print("\n--- Specific Median Values ---")
print(df[['Quantity', 'UnitPrice', 'TotalPrice', 'ItemsInCart']].median())
```

## Step 2: Identify Visual Trends Over Time
To understand the business performance, let's group the total sales (`TotalPrice`) by year and month to track historical revenue trends.

```python
# Extract Year-Month for chronological trend analysis
df['YearMonth'] = df['Date'].dt.to_period('M')

# Group revenue by month
monthly_sales = df.groupby('YearMonth')['TotalPrice'].sum().reset_index()
monthly_sales['YearMonth'] = monthly_sales['YearMonth'].astype(str) # Convert to string for clear text output

print("--- Monthly Revenue Breakdown (Top 10 Months) ---")
print(monthly_sales.sort_values(by='YearMonth').head(10))

# Quick check on the overall time range of the data
print(f"\nData spans from: {df['Date'].min().date()} to {df['Date'].max().date()}")
```

## Step 3: Spotting Outliers and Extremes
Outliers are extreme data points that deviate significantly from the rest of the distribution. We will look at maximum caps and investigate if some transactions are exceptionally massive.

```python
# Identify potential outliers using the 1.5 * IQR (Interquartile Range) rule on TotalPrice
Q1 = df['TotalPrice'].quantile(0.25)
Q3 = df['TotalPrice'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['TotalPrice'] < lower_bound) | (df['TotalPrice'] > upper_bound)]

print(f"Lower Bound Threshold for Outliers: {lower_bound}")
print(f"Upper Bound Threshold for Outliers: {upper_bound}")
print(f"Total number of outlier transactions found: {len(outliers)}")

if len(outliers) > 0:
    print("\nSample of extreme/outlier high-value orders:")
    print(outliers[['OrderID', 'Product', 'Quantity', 'TotalPrice']].head())
```

## Step 4: Summarize Operational Observations
Let's find out which products perform best, what the most frequent ordering statuses are, and where customer referrals are originating from.

```python
print("--- Top Performing Products by Volume ---")
print(df.groupby('Product')['Quantity'].sum().sort_values(ascending=False))

print("\n--- Most Common Payment Methods ---")
print(df['PaymentMethod'].value_counts())

print("\n--- Order Status Distribution ---")
print(df['OrderStatus'].value_counts())

print("\n--- Top Referral Sources ---")
print(df['ReferralSource'].value_counts())
```

### Key EDA Insights Summary:
* **Product Performance:** High volume trends will tell us what items drive baseline inventory needs.
* **Outliers:** High threshold transaction spikes indicate massive bulk orders rather than standard individual shopping carts.
* **Order Funnel:** Reviewing cancelled vs. returned vs. shipped distributions highlights logical structural leaks in orders.
